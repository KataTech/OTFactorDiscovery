```python
import sympy as sp
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

from opt import SaddlePoint
```

```python
# Define the Sympy expression for the function
x, y = sp.symbols('x y', real=True)
func = x**2 - y**2

# Create an instance of the SaddlePointOptimizer class
f = SaddlePoint(func)

# Call other methods as needed
f.optimize(90, 100, eta=0.1, max_steps=1e4, tolerance=1e-5)
f.plot(30, 30)
```
