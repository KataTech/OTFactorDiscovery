```python
import numpy as np
import sklearn.preprocessing as skpp
import sklearn.datasets as skds
import matplotlib.pyplot as plt
from kquasicenters import KQuasicenters

# Generate a synthetic dataset
np.random.seed(0)
n_samples = 200
n_features = 2
n_clusters = 2

# Generate moons
X, y = skds.make_moons(n_samples=n_samples, noise=0.05)

scaler = skpp.StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Initialize and fit KMeans
kquasi = KQuasicenters(n_clusters=n_clusters, max_iter=500, random_state=0)
kquasi.fit(X_scaled, cost_metric='euclidean', tolerance=1e-7, max_steps=200, descent_rate=0.01)

# Plot the data points and centroids
plt.scatter(X[:, 0], X[:, 1], c=kquasi.assigns, cmap='viridis')
plt.scatter(scaler.inverse_transform(kquasi.centroids)[:, 0], scaler.inverse_transform(kquasi.centroids)[:, 1], marker='X', c='red', s=100, zorder=9)
plt.title("KMeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.gca().set_aspect('equal')
plt.show()
```

```python
n_samples = 1000
X, y = skds.make_circles(n_samples=n_samples, factor=.5, noise=.05)
X = skpp.StandardScaler().fit_transform(X)

cost_metric = 'euclidean'

labels = {}
# Create a figure with 7 subplots in a row
rows = 2
plt.figure(figsize=(20, 3*rows))

for set in range(7*rows):
    stab = KQuasicenters(n_clusters=3, max_iter=900, random_state=0)
    stab.fit(X, cost_metric=cost_metric, max_steps=500)
    labels[set] = stab.assigns
    plt.subplot(rows, 7, set+1, aspect='equal')
    plt.title(f'Random Seed = {set}', size=12)  # Increase title font size
    plt.scatter(X[:, 0], X[:, 1], s=1)
    plt.scatter(stab.optimoids[:, 0], stab.optimoids[:, 1], s=50, c='red', zorder=10, marker='X')
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.xticks(())
    plt.yticks(())
plt.show()
```
