```python
import numpy as np
import sklearn.preprocessing as skpp
import sklearn.datasets as skds
import matplotlib.pyplot as plt
from kmeans import KMeans

# Generate a synthetic dataset
np.random.seed(0)
n_samples = 200
n_features = 2
n_clusters = 2

# Generate moons
X, y = skds.make_moons(n_samples=n_samples, noise=0.05)

scaler = skpp.StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Initialize and fit KMeans
kmeans = KMeans(n_clusters=n_clusters, max_iter=500)
kmeans.fit(X_scaled, cost_metric='euclidean', tolerance=1e-7, max_steps=200, learning_rate=0.01, random_state=0)

# Plot the data points and centroids
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels, cmap='viridis')
plt.scatter(scaler.inverse_transform(kmeans.centroids)[:, 0], scaler.inverse_transform(kmeans.centroids)[:, 1], marker='X', c='red', s=100, zorder=9)
plt.title("KMeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.gca().set_aspect('equal')
plt.show()
```
