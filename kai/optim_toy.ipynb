```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sympy

from src.loss import Loss
from src.optim import etgw, itgw, plot_3D, vis_gradient
```

```python
hyperbolic = Loss(lambda x, y: x**2 - y**2)
x, y, x_list, y_list, gradient_norm = etgw(hyperbolic, -10, -10, 0.1, epsilon = 0.00001, max_iter = 1000)
print("Actual Optimal Point: (0, 0)")
print(f"Estimated Optimal Point: ({x}, {y})")
vis_gradient(gradient_norm)
plot_3D(hyperbolic._loss_func(), x_list, y_list, grid = [[-10, 10], [-10, 10]], fineness = 0.2)
```

```python
sym_L = lambda x, y: (x - 1/2) * (y - 1/2) + (1/3) * sympy.exp(-(x - 1/2)**2 - (y - 3/4)**2)
eval_L = lambda x, y: (x - 1/2) * (y - 1/2) + (1/3) * np.exp(-(x - 1/2)**2 - (y - 3/4)**2)

exp_function = Loss(sym_L)
x, y, x_list, y_list, gradient_norm = etgw(exp_function, -1, -1, 0.05, epsilon = 0.00001, max_iter = 500)
print("Actual Optimal Point: (0.5, 0.5)")
print(f"Estimated Optimal Point: ({x}, {y})")
vis_gradient(gradient_norm)
plot_3D(eval_L, x_list, y_list, grid = [[-5, 5], [-5, 5]], fineness = 0.2)
```
